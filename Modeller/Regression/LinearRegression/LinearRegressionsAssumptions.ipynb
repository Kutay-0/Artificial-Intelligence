# Linear Regression Varsayımları

Linear Regression modeliyle çalışırken bazı şartların doğrulandığından emin olmalıyız. Bu şartlara "varsayımlar (assumptions)" denir.

Bu koşullar sağlandığında model bize doğru sonucu döndürür.

## Linearity
### Her x ve y arasında doğrusal ilişki bulunmalı.

Yani x değişkeni 1 birim artıyorsa y değişkeninde olan değişim her x değeri için olmalıdır.

Sağlanmaz ise model düz çizgi çekmeye çalışır. Ama veri parabolik gider.
Çözüm ise Polynomial Regression kullan ya da X'e log, karekök gibi dönüşümler uygula


## Homoscedasticity
### Residual'ların (hataların) varyansı tüm X değerleri boyunca benzer olmalı.

Yani modelin hata miktarı x küçükte olsa büyükte olsa benzer büyüklükte olmalı.
Sağlanmaz ise modelin ürettiği sonuca güven azalır.
Çözümü ise y'ye log dönüşümü veya y'ye karekök uygula.

## Multivariate Normality
### Residual'ların (hataların) normal dağılması gerekiyor.

Yani hataların büyükleri modelin bir kısmına küçükleri başka bir kısmına değilde dengeli bir şekilde bütün modele dağılması gerekiyor. Sağlanmaz ise küçük verisetlerinde güvenilmez modeller üretir. Büyük veri setlerinde kritik değildir. Çözümü ise y'ye log dönüşümü veya veri setini büyüt.

## Independece
### Her gözlem birbiriden bağımsız olmalı.

Yani bir verinin değeri, bir önceki veriden etkilenmemeli. Birbiri arasında kolerasyon olmamalı. Sağlanmaz ise tahminler sistematikleşir, değerler anlamsızlaşır. Çözümü ise Linear Regression yerine zaman serilerini kullanmak.

## Lack of Multicollinearity
### Feature'lar birbiriyle güçlü ilişki içerisinde olmamalı.

Yani 2 feature aynı şeyi anlatmamalı. Örnek üzerinden gidersek deneyim ve yaş. Yaş arttıkça deneyim artıyorsa bu istenmeyen bir durumdur. Model hangi feature'ın değer arttırdığını anlayamaz. Sağlanmaz ise katsayılar anlamsızlaşır, model tahmin yapar ama yorumlanamaz hale gelir. Çözümü ise feature'lardan birini sil, PCA ile boyut indirge ya da Ridge Regression kullan.

## Outlier Check
### Outlier, veri setindeki diğer verilerden aşırı uzak olan noktalardır.

Linear Regression outlier'lara karşı çok duyarlıdır. Çünkü OLS bu hataların karesini alır. Sağlanmaz ise Regression doğrusu kayar, R2 skoru düşer, tahminler yanıltıcı olur. Çözümü ise Outlier'ı silmek, gerçek veriyi temsil ediyorsa silme.
