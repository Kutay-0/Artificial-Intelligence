## SIMPLE LINEAR REGRESSION ŞUNU YAPAR:

### - Noktalara bakar.
### - Aralarından geçen en iyi çizgiyi bulur.
### - Çizgiyi kullanarak yeni değeri tahmin eder.

Bu çizgi şuna karşılık gelir: y = mx + b | Ev üzerinden devam edelim:

y = tahmin ettiğin değer. (Ev fiyatı)
x = verdiğin özellik.     (m^2)
m = eğim.                 (her m^2'nin etkisi)
b = sabit.                (başlangıç noktası)

### Regression'ın görevi m ve b'yi veriden öğrenmektir.

## SIMPLE LINEAR REGRESSION'IN TEMEL FORMÜLÜ

### y = b0 + b1*x + Epsilon

Epsilon hata terimidir. Gerçek hayatta veriler çizgi üzerinde durmayabilir.

O yüzden hata payı eklenir. O hata payı da epsilondur.

Epsilon = y - yt -> Bu değere "Residual" yani "Artık" denir.

Bu değerler ne kadar küçükse model o kadar başarılıdır.

## MODEL NASIL ÖĞRENİR ?
### OLS (EN KÜÇÜK KARELER)

OLS'nin amacı, tüm Residual karelerinin toplamını minimize etmektir.

SSE = Bütün (yi - yti)^2 değerlerini topla (yi = y'de i. eleman / yti = tahmin değerde i. eleman)

Kare almamızın sebebi: Negatif ve pozitifler birbirini götürmesin ve büyük hatalara daha fazla ceza verilsin.

B1 = sum((xi - x_ort)*(yi - y_ort)) / sum((xi - x_ort)^2)
B0 = y_ort - B1*x_ort

## MODELİN NE KADAR İYİ OLDUĞUNU ÖLÇME - R^2 SKORU

### Modelin başarısını ölçmek için şu formül kullanılır:

R^2 = 1 - SSE/SST (SSE = Modelin tahmin ettiği | SST = Tüm evlerin ortalama fiyatı)
